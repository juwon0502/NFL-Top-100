```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression as ln
from sklearn.linear_model import LogisticRegression as logn
from sklearn.neighbors import KNeighborsClassifier as KNN
from sklearn import tree
from sklearn.metrics import accuracy_score
from sklearn.model_selection import cross_val_score
from sklearn.tree import export_text
```

```python
wr = pd.read_csv('wr_data.csv')
rb = pd.read_csv('rb_data.csv')
wr = wr.set_index(['Year','Player'])
rb = rb.set_index(['Year','Player'])
df = pd.concat([wr,rb])
df = df.drop(columns = ['Ctch%'])
df['Top 10'] = 0
df.loc[df['Rank']<=10, 'Top 10'] = 1
y = df['Top 10']
X = df.iloc[:,4:-1]
df
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.33, random_state=0)
```

### Linear Regression

```python
lreg = ln()
lreg.fit(X_train, y_train)
y_pred = lreg.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error
mean_squared_error(y_test, y_pred)
```

### KNN

```python
df['Top 10'] = 0
df.loc[df['Rank']<=10, 'Top 10'] = 1
tens = df.loc[df['Top 10']==1]
y = df['Top 10']
```
